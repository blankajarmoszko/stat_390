```python
%matplotlib inline
import pandas as pd 
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.tools.eval_measures import rmse
from sklearn.metrics import mean_squared_error
from math import sqrt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pandas.plotting import autocorrelation_plot
from mango import Tuner
import itertools
from pmdarima.arima import auto_arima

# Suppress all warnings
import warnings
warnings.simplefilter("ignore")
```

# Feature engineering

```python
# Predictive variable
pred_var = 'Additional_Doses_Vax_Pct'

# Import data with specific columns
df = pd.read_csv('df_post_eda.csv', usecols=['Date', pred_var])

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Group by 'Date' to ensure there is one unique value per date
df = df.groupby('Date').mean()

# Sorting the DataFrame by the index just to make sure it's in chronological order
df.sort_index(inplace=True)

# Now 'Date' is the index and I have the average of 'Additional_Doses_Vax_Pct' per date
```

# Test & train

```python
# Calculate the index for the 70-30 split
split_index = int(len(df) * 0.7)

# Split into train and test
train = df.iloc[:split_index][pred_var]
test = df.iloc[split_index:][pred_var]

# Check the size of each dataset
print(f"Training set size: {len(train)}, Test set size: {len(test)}")
```

# Model 1

```python
# Run auto_arima to find optimal SARIMA model parameters
model = auto_arima(train, start_p=0, start_q=0,
                        max_p=3, max_q=3, m=12,
                        start_P=0, seasonal=True,
                        d=1, D=1, trace=True,
                        error_action='ignore',  
                        suppress_warnings=True, 
                        stepwise=True)

print(model.summary())
print(f"AIC: {model.aic()}")
```

```python
# Train results
train_pred=model.fittedvalues()
pd.concat([train, train_pred], axis=1).plot(figsize=(20,8))
plt.legend(['train', 'train_pred'])

# Calculate and print the RMSE
train_rmse = rmse(train, train_pred)
print(f"Train RMSE: {train_rmse}")
```

```python
# Test results
test_pred=model.predict(n_periods=len(test))
pd.concat([test, test_pred], axis=1).plot()

plt.legend(['test', 'test_pred'])

# Calculate and print the RMSE
test_rmse = rmse(test, test_pred)
print(f"Test RMSE: {test_rmse}")
```

# Model 2
To improve the model, here are the steps I am considering:

1. Simplify the Model: Reduce the complexity of the model by limiting the number of moving average terms. The Ljung-Box test indicates that there might be some structure in the residuals that is not captured by the model, but the high order of MA terms does not seem to be addressing this effectively.

2. Address Overfitting: Since the model is not generalizing well to the test set, consider implementing cross-validation to better tune the model parameters.

3. Seasonality Adjustment: Given the seasonal nature of the data, make sure that the seasonality is adequately captured. The SARIMAX(2,1,0)(0,1,0)[12] model suggests that annual seasonality is considered, but may need to refine how the seasonal terms are modeled.

```python
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error
from math import sqrt

# Assuming 'train' and 'test' are your training and test sets respectively
# and pred_var is the column with the values you want to predict

# Initialize variables to store the best score and model
best_rmse = float("inf")
best_model = None
best_m = None

# Trying different seasonal periods
for m in [6, 12, 24]:
    temp_model = auto_arima(train,
                            start_p=0, start_q=0,
                            max_p=5, max_q=5,
                            m=m,              # Testing different m values
                            start_P=0, start_Q=0,
                            max_P=3, max_Q=3,
                            seasonal=True,    # Seasonality is expected
                            d=1, D=1,         # Allowing model to differentiate
                            trace=True,
                            error_action='ignore',  
                            suppress_warnings=True, 
                            stepwise=True)

    # Print out the summary of the fit model
    print(f"Model summary for m={m}:")
    print(temp_model.summary())

    # Predict on the test set
    test_pred = temp_model.predict(n_periods=len(test))

    # Calculate the RMSE for the test set
    test_rmse = sqrt(mean_squared_error(test, test_pred))
    print(f'Test RMSE for m={m}: {test_rmse}')

    # If the test RMSE for the current model is the best, store the model
    if test_rmse < best_rmse:
        best_rmse = test_rmse
        best_model = temp_model
        best_m = m

# After the loop, print the best RMSE and corresponding m value
print(f'Best Test RMSE: {best_rmse} for seasonal period m={best_m}')
```

```python
# Run auto_arima to find optimal SARIMA model parameters
model = auto_arima(train, start_p=0, start_q=0,
                        max_p=10, max_q=10, m=24,
                        start_P=0, start_Q=0,
                        max_P=10, max_Q=10,seasonal=True,
                        d=1, D=1, trace=True,
                        error_action='ignore',  
                        suppress_warnings=True, 
                        stepwise=True)

print(model.summary())
print(f"AIC: {model.aic()}")
```

```python
# Train results
train_pred=model.fittedvalues()
pd.concat([train, train_pred], axis=1).plot(figsize=(20,8))
plt.legend(['train', 'train_pred'])

# Calculate and print the RMSE
train_rmse = rmse(train, train_pred)
print(f"Train RMSE: {train_rmse}")
```

```python
# Predict on the test set
test_pred = model.predict(n_periods=len(test))
pd.concat([test, test_pred], axis=1).plot()

plt.legend(['test', 'test_pred'])

# Calculate the RMSE for the test set
test_rmse = sqrt(mean_squared_error(test, test_pred))
print(f"Test RMSE: {test_rmse}")
```
